```python
from google.colab import files
import pandas as pd

d1 = files.upload()
icd9_df = pd.read_csv('ICD9-430_Patients_DischargeSummary_Random.csv')
```

```python
# Using en_core_sci_md library
import en_core_sci_md
nlp = en_core_sci_md.load()
icd9_df["sci_spacy_doc"] = icd9_df["TEXT"].apply(nlp)

for doc in icd9_df["sci_spacy_doc"].head(2):
    print("Entities in Text:", [(ent.text, ent.start_char, ent.end_char, ent.label_) for ent in doc.ents])
    print("--------------------------------------------------")
    for token in doc:
      print(token.text)

print("********************Entity Visualizer*******************************")

# Entity Visualizer
from spacy import displacy
for doc in icd9_df["sci_spacy_doc"]:
    displacy.render(doc, style="ent", jupyter=True)
    print("***************************************************")
```
